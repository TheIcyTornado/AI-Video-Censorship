```python
# level 1 - Basic Profanity Censorship

import pandas as pd, string, re

# --- Load the CSV file ---
profanity_df = pd.read_csv("profanity.csv")

# --- Normalize columns ---
profanity_df["text"] = profanity_df["text"].str.lower()
profanity_df["severity_description"] = profanity_df["severity_description"].str.lower()

# --- Create sets of swear words ---

low_swear_words = set(profanity_df[profanity_df["severity_description"] == "mild","strong","severe"]["text"])
mid_swear_words = set(profanity_df[profanity_df["severity_description"] == "strong","severe"]["text"])
strong_swear_words = set(profanity_df[profanity_df["severity_description"] == "severe"]["text"])

swear_words = mid_swear_words  # Change this to low_swear_words or strong_swear_words as needed

sample_text = """
I have had it with these motherfucking snakes on this motherfucking plane!
"""

def censor_swear_words(text):
    def censor_match(match):
        word = match.group()
        lower_word = re.sub(rf"[{string.punctuation}]", "", word).lower()
        if lower_word in swear_words:
            # Keep punctuation but replace letters
            return re.sub(r"[A-Za-z]", "*", word)
        return word

    # Replace word by word while keeping punctuation
    return re.sub(r"\b[\w'-]+\b|[^\w\s]", censor_match, text)

# --- Run censoring ---
censored_text = censor_swear_words(sample_text)

print("Original Text:", sample_text)
print("Censored Text:", censored_text)
```

```python
# level 2 - Video Transcription with Profanity Censorship

import os,whisper

# --- Config ---
video_path = r"C:\Users\jjhip\OneDrive - University of Wolverhampton\MCS\MCS Project\Project\OTHER\Snakes.mp4"
Video_Name = os.path.splitext(os.path.basename(video_path))[0]
model_size = "base"

# --- Load Whisper and transcribe ---
model = whisper.load_model(model_size)
result = model.transcribe(video_path)
text = result["text"].strip()

# --- Save transcription ---
with open(f"{Video_Name}_transcription.txt", "w", encoding="utf-8") as f:
    f.write(text)
print("Transcription Preview:\n", text, "\n")

# --- Censor transcription ---
censored_text = censor_swear_words(text)
with open(f"{Video_Name}_censored_transcription.txt", "w", encoding="utf-8") as f:
    f.write(censored_text)
print("Censored Transcription Preview:\n", censored_text)

```
